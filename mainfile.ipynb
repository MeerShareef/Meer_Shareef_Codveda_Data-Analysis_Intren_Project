```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("1) iris.csv")  # Adjust path if needed

# Drop duplicates and standardize species
df = df.drop_duplicates()
df['species'] = df['species'].str.strip().str.lower()

# === BAR PLOT: Average Petal Length by Species ===
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='species', y='petal_length')
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.tight_layout()
plt.savefig('bar_plot_petal_length_by_species.png')  # Saves image

# === LINE CHART: Petal Width Trend over Samples ===
plt.figure(figsize=(8, 5))
for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.plot(subset.index, subset['petal_width'], label=species)

plt.title('Petal Width Trend by Species')
plt.xlabel('Sample Index')
plt.ylabel('Petal Width (cm)')
plt.legend()
plt.tight_layout()
plt.savefig('line_chart_petal_width_by_species.png')

# === SCATTER PLOT: Sepal Length vs Sepal Width ===
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.tight_layout()
plt.savefig('scatter_plot_sepal.png')

# Show all plots (optional)
plt.show()
```
